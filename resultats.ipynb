```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
col_names = ["N=2","N=3","N=4","N=5","N=6"]
index_names = ["N=10","N=50","N=100","N=500","N=1000","N=5000","N=10000","N=50000","N=100000"]
df_standard = pd.DataFrame([[6,7,8,9,9],
                           [11,16,23,30,36],
                           [13,21,32,45,58],
                           [18,31,54,88,136],
                           [20,35,63,112,181],
                           [25,45,87,167,304],
                           [27,49,96,191,370],
                           [31,57,118,251,511],
                           [33,61,127,274,586]],
                          columns = col_names,index=index_names)
df_standard
```

```python
fig = plt.figure(figsize=(6,3),dpi=150,)

axes = fig.add_axes([0,0,1,1])
axes.plot(df_standard["N=2"],label=["N=2"])
axes.plot(df_standard["N=3"],label=["N=3"])
axes.plot(df_standard["N=4"],label=["N=4"])
axes.plot(df_standard["N=5"],label=["N=5"])
axes.plot(df_standard["N=6"],label=["N=6"])

axes.tick_params(axis='both', labelsize=8)

plt.title("k-d Standard Binary-Tree")

plt.xlabel("# nodes de l'arbre",fontsize=10)
plt.ylabel("# nodes consultats durant la cerca", fontsize=10)
plt.legend()

plt.show()
```
